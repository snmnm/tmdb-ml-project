```python
import pandas as pd

df = pd.read_csv('/content/tmdb_5000_movies.csv')
df.head()
```

Hangi sütunlarda ne kadar eksik veri olduğunu inceliyoruz


```python
df.isnull().sum()
```

```python
cols_to_drop = [
    "homepage",
    "overview",
    "tagline",
    "status",          # model için anlamsız
    "original_title",  # title ile aynı
    "original_language",
    "spoken_languages",
    "production_countries",
    "production_companies"
]

df = df.drop(columns=cols_to_drop, errors="ignore")
df.head()
```

# Release date'i release year çevirdik

```python
df["release_year"] = pd.to_datetime(df["release_date"], errors="coerce").dt.year
df = df.drop(columns=["release_date"])
```

# Runtime daki iki eksik değeri median ile doldurduk

```python
df["runtime"] = df["runtime"].fillna(df["runtime"].median())
```

# JSON kolonları tekrar düzgün listeye çevirelim

```python
import ast

json_cols = ["genres", "keywords"]

def parse_json_to_list(x):
    # 1) Eğer zaten liste ise direkt geri dön
    if isinstance(x, list):
        return x

    # 2) Boş / NaN / bos string durumları
    try:
        if pd.isna(x) or x == "" or x == "[]":
            return []
    except TypeError:
        # pd.isna listelerde array döndürmesin diye
        return []

    # 3) String ise JSON gibi parse etmeyi dene
    try:
        parsed = ast.literal_eval(x) if isinstance(x, str) else x
        if isinstance(parsed, list):
            return parsed
        else:
            return []
    except Exception:
        return []

# Kolonlara uygula
for col in json_cols:
    df[col] = df[col].apply(parse_json_to_list)

# Liste uzunluklarından sayısal feature çıkar
df["genre_count"] = df["genres"].apply(len)
df["keyword_count"] = df["keywords"].apply(len)

```

## **Sayısal değişken histogramları**
Grafikler;
bütçe dağılımı,
gelir dağılımı,
popülerlik dağılımı,
oy sayısı dağılımı,

```python
import matplotlib.pyplot as plt  #matplotlib: klasik grafik çizme
import seaborn as sns            #seaborn: estetik grafikler için

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns #df icinde sadece sayısal sütunları seciyor

plt.figure(figsize=(15, 30))    #grafiklerin büyük ve okunaklı olması için tuvali ayarlıyoruz, 15x30

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(8, 3, i)
    sns.histplot(df[col], kde=True, color='skyblue')    #histogram ciziyor
    plt.title(col)       #her grafiğe adını giriyor

plt.tight_layout()
plt.show()         # her grafiği ekrana bastırıyor
```
